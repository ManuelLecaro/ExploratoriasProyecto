```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasetDepurado.csv")
df = df.drop_duplicates(subset=['owner_id','title'])
df.head()
```

```python
# Agrupando por ciudad
ciudadGroupDf = df[['location','only_price','reviews']].groupby('location').mean()
ciudadGroupDf.head()
```

```python
# Unión de datasets para obtener densidad poblacional de la ciudad a la que el alojamineto pertenece

# Carga de los datasets auxiliares
provCantonesDf = pd.read_csv("cantonesPorProvincia.csv", encoding="utf-8")
datosProvinciaDf = pd.read_csv("datosPoblacion.csv", encoding="utf-8")
hogaresInternetDf = pd.read_csv("poblacionPorProvincia.csv", encoding="utf-8")

# Pre procesamiento
provCantonesDf["Provincia"] = provCantonesDf["Provincia"].str.title() # Convirtiendo texto de mayúsculas a formato título (primera letra mayúscula)
provCantonesDf["Cantón"] = provCantonesDf["Cantón"].str.title()
provCantonesDf["Región"] = provCantonesDf["Región"].str.title()
datosProvinciaDf["Provincia"] = datosProvinciaDf["Provincia"].str.title() # Convirtiendo texto de mayúsculas a formato título (primera letra mayúscula)
datosProvinciaDf["Cantón"] = datosProvinciaDf["Cantón"].str.title()

# Unión
alojamientosRegionesDf = ciudadGroupDf.merge(provCantonesDf, left_on="location", right_on="Cantón")
alojamientosPoblaciónDf = alojamientosRegionesDf.merge(datosProvinciaDf, left_on="Cantón", right_on="Cantón")
finalDf = alojamientosPoblaciónDf.merge(hogaresInternetDf, left_on="Provincia_x", right_on="provincia")
finalDf.head()
```

```python
# Identificando las variables numéricas
features = ['only_price','reviews','Densidad']

# Aislando las variables
x = finalDf.loc[:, features]
# Estandarizando las variables
x = StandardScaler().fit_transform(x)
x = pd.DataFrame(data=x, columns=['precio','reviews','densidad'])
x.head()
```

```python
# Encontrando valores aberrantes
outliers = x[(x['reviews']>3)|(x['reviews']<-3)|(x['densidad']>3)|(x['densidad']<-3)|(x['precio']>3)|(x['precio']<-3)]
print(outliers)
# Removiendo dichos valores del dataframe original y del estandarizado
finalDf = finalDf.drop(list(outliers.index.values))
x = x.drop(list(outliers.index.values))
```

```python
import matplotlib as mp
from mpl_toolkits.mplot3d import axes3d, Axes3D 
fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(45,45)
ax.set_title('Airbnb ciudades', fontsize = 20)
ax.scatter3D(x['precio'],x['reviews'],x['densidad'])
#ax.legend(["not infected", "infected"])
# label the axes
ax.set_xlabel("precio")
ax.set_ylabel("reviews")
ax.set_zlabel("densidad")
```

```python
# Metodo del codo
distortions = []
for i in range(1, 11):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(x)
    distortions.append(km.inertia_)

plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Número de clusters')
plt.ylabel('Distorción')
plt.show()
```

```python
#Clustering

kmeans = KMeans(n_clusters=3).fit(x)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(45,45)
ax.set_title('Airbnb ciudades', fontsize = 20)
ax.scatter3D(x['precio'],x['reviews'],x['densidad'], c=kmeans.labels_)
# label the axes
ax.set_xlabel("precio")
ax.set_ylabel("reviews")
ax.set_zlabel("densidad")
```

```python
print(len(kmeans.labels_))
```

```python
# Agregando categoría a las ciudades según los clusters
finalDf['cluster'] = kmeans.labels_
finalDf.head()
```

```python
# Agrupando
finalDf[['cluster','reviews','only_price','Densidad']].groupby('cluster').mean()
```
