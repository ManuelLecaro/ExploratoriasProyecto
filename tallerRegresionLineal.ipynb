### Pregunta 1: humedad vs temperatura

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
```

```python
df = pd.read_csv("weatherHistory.csv")
```

```python
y=df["Temperature (C)"]
x=df["Humidity"]
y = y.values.reshape(len(y),1)
x = x.values.reshape(len(x),1)

x_train = x[:-250]
x_test = x[-250:]
y_train = y[:-250]
y_test = y[-250:]
```

```python
plt.scatter(x_test, y_test,  color='black')
plt.title('Test Data')
plt.xlabel('humidity')
plt.ylabel('temperature')
plt.show()
```

Se observa claramente que los puntos se agrupan formando un patron decreciente, por lo que es posible que exista una correlacion entre ambas variables

### Pregunta 2: humedad vs temperatura aparente

```python
y=df["Apparent Temperature (C)"]
x=df["Humidity"]
y = y.values.reshape(len(y),1)
x = x.values.reshape(len(x),1)

x_train = x[:-250]
x_test = x[-250:]
y_train = y[:-250]
y_test = y[-250:]
```

### Pregunta 3: humedad vs temperatura aparente

```python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(x_train, y_train)

# Plot outputs
plt.title('Test Data')
plt.xlabel('humidity')
plt.ylabel('apparent temperature')
y_predict = regr.predict(x_test)
plt.scatter(x_test, y_test,  color='black')
plt.plot(x_test, y_predict, color='green',linewidth=3)
plt.show()
```

La recta calculada por regresi√≥n hace un buen trabajo agrupando a su alrededor los puntos del gr√°fico de dispersi√≥n, aunque se ve sesgada hacia los datos con los cuales ha sido entrenado (el 75% de los datos que fueron parte del set de training x_train y y_train). Sin embargo comprobaremos que en efecto este modelo puede predecir con ciert precision resultados en test.

```python
# Resultados
print("Intercepto: %.2f" % regr.intercept_)
print('Coeficiente en x: %.2f' %regr.coef_[0][0])
print("Suma cuadr√°tica del error: %.2f"
      % mean_squared_error(y_test, y_predict))
# Varainza R cuadrado: un resultado de 1 seria una predicion perfecta
print('Varianza: %.2f' % r2_score(y_test, y_predict))
```
